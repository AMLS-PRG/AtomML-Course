```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
file_path = "AtomML-Course/module-5/03-Trained MLPs (Folder 02 Output)/lcurve.out"
```

```python

colnames = [
    "step", "rmse_val", "rmse_trn", "rmse_e_val", "rmse_e_trn",
    "rmse_f_val", "rmse_f_trn", "lr"
]


df = pd.read_csv(
    file_path,
    delim_whitespace=True,
    comment="#",
    names=colnames,
    engine="python"
)


df = df.astype(float)

plt.figure(figsize=(14, 6))

# ------- Energy RMSE Plot -------
plt.subplot(1, 2, 1)
plt.plot(df["step"], df["rmse_e_val"], label="Validation", color="tab:blue")
plt.plot(df["step"], df["rmse_e_trn"], label="Train", color="tab:orange")
plt.yscale("log")
plt.xlabel("Training Step")
plt.ylabel("Energy RMSE (log scale)")
plt.title("Log-scale Energy RMSE")
plt.grid(True, which="both", ls="--", alpha=0.5)
plt.legend()

# ------- Force RMSE Plot -------
plt.subplot(1, 2, 2)
plt.plot(df["step"], df["rmse_f_val"], label="Validation", color="tab:green")
plt.plot(df["step"], df["rmse_f_trn"], label="Train", color="tab:red")
plt.yscale("log")
plt.xlabel("Training Step")
plt.ylabel("Force RMSE (log scale)")
plt.title("Log-scale Force RMSE")
plt.grid(True, which="both", ls="--", alpha=0.5)
plt.legend()

plt.tight_layout()
plt.show()
```
