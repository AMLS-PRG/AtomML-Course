```python
!pip install ase numpy

import ase.io as aseio
import numpy as np
from ase.visualize import view
from ase.io.lammpsdata import read_lammps_data, write_lammps_data
from ase.build import make_supercell
from google.colab import files
```

```python
file_path = "AtomML-Course/module-5/04-Performing MD simulations/small-data/si-64-liquid.lammps-data"
```

```python

atoms = read_lammps_data(file_path, style='atomic')

P = np.identity(3) * 3
supercell = make_supercell(atoms, P)

write_lammps_data('liquid_supercell_3x3x3.data', supercell)
# Read a single frame (change index to view other frames)
atoms = aseio.read('liquid_supercell_3x3x3.data', format='lammps-data', style='atomic')

# Replace element labels if needed
symbols = np.array(atoms.get_chemical_symbols())
symbols[symbols == 'H'] = 'Si'
atoms.set_chemical_symbols(symbols)

# Output information
print(f"Number of atoms: {len(atoms)}")
print("Cell vectors (box size) in Å:")
print(atoms.get_cell())

files.download('liquid_supercell_3x3x3.data')

# Show 3D structure
view(atoms, viewer='x3d')

```
